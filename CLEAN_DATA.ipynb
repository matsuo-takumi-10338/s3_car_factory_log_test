### Unity Catalogテーブル定義/S3テーブルデータ削除

```python
database_bronze_path = "s3://databricks-sample-bucket-20260602/car_factory_dev/database/bronze/"
env = "dev"
table_name = f"car_factory_{env}.bronze.car_factory_bronze"

try:
    items = dbutils.fs.ls(database_bronze_path)
    for item in items:
        dbutils.fs.rm(item.path, recurse=True)
    print("SUCCESS: DELETE TABLE DATA")
except Exception as e:
    if "CloudFileNotFoundException" in str(e) or "java.io.FileNotFoundException" in str(e):
        print("対象のフォルダは既に空か、存在しません")
    else:
        print(f"ERROR:{e}")

try:
    spark.sql(f"DROP TABLE IF EXISTS {table_name}")
    print("SUCCESS: DROP TABLE")
except Exception as e:
    print(f"ERROR:{e}")
```
