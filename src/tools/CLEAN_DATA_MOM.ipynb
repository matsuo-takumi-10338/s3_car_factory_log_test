### 設定値

### S3 Rawデータの削除

```python
try:
    files = dbutils.fs.ls("s3://cf-mom-factory-bucket-dev/mom_factory_dev/raw/")
    for file in files:
        dbutils.fs.rm(file.path, recurse=True)
    print("SUCCESS")
except Exception as e:
    print(f"ERROR:{e}")
```

### S3 チェックポイント・スキーマの削除

```python
checkpoint_path = f"s3://cf-mom-factory-bucket-dev/mom_factory_dev/checkpoints/"

try:
    checkpoints = dbutils.fs.ls(checkpoint_path)
    for cp in checkpoints:
        dbutils.fs.rm(cp.path, recurse=True)
    print("SUCCESS")
except Exception as e:
    print(f"ERROR:{e}")
```
